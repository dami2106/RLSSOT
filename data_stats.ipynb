```python
import glob
from collections import Counter
import matplotlib.pyplot as plt
```

```python
file_list = glob.glob('examples/data/desktop_assembly/groundTruth/*')
```

```python
action_counter = Counter()
```

```python
for filename in file_list:
    with open(filename, "r") as file:
        for line in file:
            action = line.strip()  # Remove any extra whitespace/newlines
            if action:  # Ignore empty lines
                action_counter[action] += 1
```

```python
actions = list(action_counter.keys())
counts = list(action_counter.values())
```

```python
plt.figure(figsize=(10, 6))
plt.bar(actions, counts, color='skyblue')
plt.xlabel("Actions")
plt.ylabel("Counts")
plt.title("Distribution of Actions")
plt.xticks(rotation=45)  # Rotate labels if needed for better readability
plt.tight_layout()
plt.show()
```
